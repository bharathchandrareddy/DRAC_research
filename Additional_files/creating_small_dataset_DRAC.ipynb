from the n number of files present in each disaster name we need to copy 20 files from each disaster name to new dataset.

```python
import os
import shutil


dataset_path = '/content/drive/MyDrive/disaster_management_data/train/images'
destination_path = '/content/drive/MyDrive/XBD/train/images'
label_path = '/content/drive/MyDrive/disaster_management_data/train/labels'
destination_label_path = '/content/drive/MyDrive/XBD/train/labels'
for disaster_name in df['disaster_name'].unique():
  print(disaster_name)
  disaster_df=df[df['disaster_name']==disaster_name].head(20)
  for index, row in disaster_df.iterrows():
        pre_image_name = row['pre_image_name']
        post_image_name = row['post_image_name']

        pre_image_path = os.path.join(dataset_path, pre_image_name)
        post_image_path = os.path.join(dataset_path, post_image_name)

        #for pre_disaster image
        if os.path.exists(pre_image_path):
            shutil.copy2(pre_image_path, destination_path)
            print(f'copied :{pre_image_name} to destination {destination_path}')
        #for post disaster image
        if os.path.exists(post_image_path):
            shutil.copy2(post_image_path, destination_path)
            print(f'copied {post_image_name} to destination {destination_path}')

        #for per_disaster json files
        pre_json_path = os.path.join(label_path, pre_image_name[:-4] + '.json')
        if os.path.exists(pre_json_path):
            shutil.copy2(pre_json_path, destination_label_path)
            print(f'copied {pre_json_path} to destination {destination_label_path}')
        #for post_disaster_json files
        post_json_path = os.path.join(label_path, post_image_name[:-4] + '.json')
        if os.path.exists(post_json_path):
            shutil.copy2(post_json_path, destination_label_path)
            print(f'copied {post_json_path} to destination {destination_label_path}')
print("Images copied successfully.")
```
